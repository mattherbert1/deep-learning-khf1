## Exercise 2:

```python
rgb_array = list(itertools.chain(*rgb_array)) # flattening rgb_array so that it is containing only RGB tuples of all of the pictures
```

Calculating mean and standard deviation:

```python
rgb_array_mean = np.mean(rgb_array, axis=0) # calculating mean values per channel, creating a vector
print(f'mean:\n R: {rgb_array_mean[0]} G: {rgb_array_mean[1]} B: {rgb_array_mean[2]}')
rgb_array_std = np.std(rgb_array, axis=0) # calculating standard deviation per channel, creating a vector
print(f'standard deviation:\n R: {rgb_array_std[0]} G: {rgb_array_std[1]} B: {rgb_array_std[2]}')
```

Transforming dataset:

```python
rgb_array = (rgb_array - rgb_array_mean) / rgb_array_std # transforming dataset to 0 mean and 1 standard deviation (using numpy's array broadcasting)
```

Checking mean and standard deviation of transformed dataset:

```python
rgb_array_mean = np.mean(rgb_array, axis=0) # calculating mean values per channel of the transformed dataset
print(rgb_array_mean) # printing mean vector (which is approximately [0 0 0] with low inaccuracy)
rgb_array_std = np.std(rgb_array, axis=0) # calculating standard deviation values per channel of the transformed dataset
print(rgb_array_std) # printing standard deviation vector (which is approximately [1 1 1] with low inaccuracy)
```

## Exercise 3:

```python
sample_rate, samples = wavfile.read('sitar.wav') # reading sound file, getting sample rate and sample chunks
# creating plot
plt.specgram(samples, Fs=sample_rate) # creating spectrogram
plt.ylabel('Frequency [Hz]') # setting label of Y axis
plt.xlabel('Time [sec]') # setting label of X axis
plt.show() # displaying plot
```

## Exercise 4:

```python
page = wikipedia.page("Deep learning") # loading WikiPage with title "Deep learning"
print(page.content) # printing the plain text content of the page
letters_dict = {} # creating dictionary for the letters and their numbers of occurrences
for i in page.content: # iterating through the text content of the page
    if not str.isalpha(i): # if the character is not a letter then it should not be counted
        continue
    i = str.lower(i) # transforming the letter to lower case
    if not (i in letters_dict.keys()): # if the letter has not been in the dictionary yet, then it will be put into that
        letters_dict[i] = 1 # this is the first occurrence of the letter
    else:
        letters_dict[i] += 1 # increasing the occurrence of the letter by 1
# sorting the letters in reverse order by the number of the occurrence
letters_list = [(k, v) for k, v in letters_dict.items()] # putting key-value pairs into an array of tuples
sorted_letters_list = sorted(letters_list, key=lambda x: x[1], reverse=True) # sorting the list in reverse order of the occurrences
# separating keys and values into two arrays
keys = [x for (x, _) in sorted_letters_list]
values = [y for (_, y) in sorted_letters_list]
plt.bar(keys, values) # creating plot
plt.show() # displaying plot
```
